# **Substituir valores de forma condicional**

Existe momentos que precisamos substituir os valores da coluna de forma condicional. Eu vi vários cursos ensinando criar colunas extras, depois exclui a coluna antiga, criando várias etapas desnecssárias.

Segue abaixo a forma de substituir de forma condicional

1) Selecione a coluna desejada e com botão direito, ir na opção substituir.
2) Nessa parte, coloque qualquer valor, apenas para criar a etapa
3) Agora com a etapa criada, é só alterar usando each if.

Segue um exemplo na parte de baixo

```python
# Substitui o valor 1 por 0 da coluna ID Projeto, apenas para criar essa linha

= Table.ReplaceValue("Tipo_Alterado",1,0,Replacer.ReplaceValue,{"ID Projeto"})
```

```python
#  Agora fiz essas substituições, usando if

= Table.ReplaceValue(
"Tipo_Alterado",
each [ID Projeto], # Esse each é para percorrer na coluna ID Projeto
each if [ID Projeto] = null then [Conta Contábil] # Se ID Projeto for = nulo, então pegue o valor da Conta Contábil
else if [ID Projeto] = 45577 then 53222 # Se ID Projeto for igual = 45577 então é 53222
else if [Conta Contábil] = null then [ID Projeto] / 2 # Se Conta Contábil for nulo, então Divide ID Projeto por 2
else [ID Projeto],Replacer.ReplaceValue,{"ID Projeto"}) # Por último, pegue ID do Projeto
```
