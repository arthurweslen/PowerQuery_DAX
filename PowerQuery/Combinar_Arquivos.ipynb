## Combinar Arquivos da Mesma Pasta

É comum em algumas empresas ter arquivos mensais, então existe a necessidade de combinar os arquivos dentro de um diretório. Existe o módo básico,simples que cria umas tabelas desnecessárias e uma avançada.

### 1) O Módo Básico

• Primeiro conecta na Pasta onde estão os arquivos <br>
• Depois ira aparecer uma tabela com todos os arquivos <br>
• Na coluna **Content**, clica no ícone *(conforme a imagem abaixo)*

Isso é simples, mas polui muito o PowerBI *(Tabelas de transformações)*

### 1) O Módo AVANÇADO

• Ao invés de clicar no ícone da coluna **content**, vamos criar uma coluna personalizada *(vou chamar de dados)*  <br>

> =Excel.Workbook([Content])


• Clique para expandir todas as colunas dessa coluna nova *(principalmente data e Kind)* <br>
• Filtre apenas as linhas que contem os conteúdos, no meu caso, foi usar Kind = Sheet <br>
• Deixe apenas a coluna dados e depois expanda os arquivos
