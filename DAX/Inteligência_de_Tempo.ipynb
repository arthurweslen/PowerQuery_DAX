#Inteligência de Tempo (Comparação Anual)

####Diferença do **SAMEPERIODLASTYEAR** e **PREVIOUSYEAR**

***• SAMEPERIODLASTYEAR***: Traz o mesmo período do ano passado, respeitando o nível de detalhe (dia, mês, trimestre). <br>
***• PREVIOUSYEAR***: Traz o ano anterior inteiro, independentemente do filtro. *(Ignora a granularidade de Trimestre, Mês e dia)*

####Cenário

Mostre o valor totaldo ano passado usando as duas medidas

```python
Valor Ano Passado SAMEPERIODLASTYEAR =
CALCULATE(
    SUM(Tabela[Valor]),
    SAMEPERIODLASTYEAR(d_Calendario[Date]))
```

```python
Valor Ano Passado PREVIOUSYEAR =
CALCULATE(
    SUM(Tabela[Valor]),
    PREVIOUSYEAR(d_Calendario[Date]))
```

### Variação Anual

Para esse exemplo, utilizaremos a medida [Valor Ano Passado SAMEPERIODLASTYEAR], pois nesse cenário queremos comparar com o o mesmo mês do ano passado


```python

% Variação Anual =
VAR Valor_AnoAnterior = [Valor Ano Passado SAMEPERIODLASTYEAR]
VAR Valor_AnoAtual = SUM(Tabela[Valor])

VAR Diferenca = Valor_AnoAtual - Valor_AnoAnterior
VAR Percentual = DIVIDE(Diferenca,Valor_AnoAnterior)

RETURN Percentual
```

#Inteligência de Tempo (Comparação Mensal)

####Diferença do **DATEADD** e **PREVIOUSMONTH**
<br>

**• DATEADD**(-1, MONTH) respeita o intervalo atual e volta exatamente um mês.
<br>
**• PREVIOUSMONTH** sempre retorna o mês anterior completo. *(Ignora a granularidade de dia)*

```python
Valor Mes Anterior PREVIOUSMONTH =
CALCULATE(
    SUM(Tabela[Valor]),
    PREVIOUSMONTH(d_Calendario[Date]))
```

```python
Valor Mês Anteior DATEADD =
CALCULATE(
    SUM(Tabela[Valor]),
    DATEADD(d_Calendario[Date],-1,MONTH))
```

### Variação Mensal

Para esse exemplo, utilizaremos a medida [Valor Mês Anteior DATEADD].


```python
% Variação Mensal =
VAR Valor_MesAnterior = [Valor Mês Anteior DATEADD]
VAR Valor_MesAtual = SUM(Tabela[Valor])

VAR Diferenca = Valor_MesAtual - Valor_MesAnterior
VAR Percentual = DIVIDE(Diferenca,Valor_MesAnterior)

RETURN Percentual
```
