#**USERELATIONSHIP**
*(Relacionmentos Inativos)*


**USERELATIONSHIP** é uma função do DAX usada dentro do CALCULATE para ativar uma relação inativa entre duas tabelas, apenas naquele cálculo específico.

No exemplo abaixo vamos utilizar a tabela que relacionao a data de pagamento com a d_calendario. Porém a data de recebimento dos funcionários é diferente do dia do pagamento que a empresa faz para a empresa terceirizada. A solução é criar um relacionamento Inativo e usar o *USERELATIONSHIP*


1) Crie um relacionamento ativo com a d_Calendario
2) Crie um segundo relacionamento na d_Calendario, porém inativo *(pois não pode relacionar mais de uma coluna com as mesmas tabelas)*
3) Crie a medida




```python
Valor Recebido (Terceirizada) =
CALCULATE(
    [Total Pagamento],
    USERELATIONSHIP('f_Pagamento_Recebimento'[Data Envio],d_Calendario[Date]))
```
