# % Percentual sobre Categoria e subcategoria

Em alguns casos, principalmente visuais de matriz, que expande para um segundo nível, é interessante ele fazer o percentual. No exemplo abaixo eu tenho a Categoria *(Primeiro Nível)*: Continente e a Subcategoria *(Segundo Nível)* com o País

A Medida ficará assim:

1) Crie uma variável com o Valor Total Vendido ou reutilize uma medida *(um simples SUM)*
2) Crie um calculate com o Valor TOtal Vendido, mas combinando all(tabela) e Values (Tabela[Coluna]) onde essa coluna seria a categoria ou o nível 1 da sua matriz
3) Crie uma variável para realizar a divisão e retorne ela
4) Coloque no formato Percentual
